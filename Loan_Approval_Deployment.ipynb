```python
import pandas as pd
import numpy as np
import sklearn
import seaborn as sns 
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split , GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
import pickle
```

```python
print('numpy' , np.__version__)
print('pandas' , pd.__version__)
print('sklearn' , sklearn.__version__)
```

```python
data = pd.read_csv(r"C:\Users\TheUnseenVlogs\Downloads\Data Analysis Project\Loan_Approval_Deployment\credit_risk_dataset.csv")
data
```

```python
data = data.dropna()
data
```

```python
data['loan_intent_duplicate'] = data['loan_intent'].apply(lambda x:0 if x == 'EDUCATION' else 1  if x == 'MEDICAL' else 2 if x =='VENTURE' else 3 if x == 'PERSONAL' else 4 if x == 'DEBTCONSOLIDATION' else 5)
data['loan_grade_duplicate'] = data['loan_grade'].apply(lambda x:0 if x == 'A' else 1  if x == 'B' else 2 if x=='C' else 3 if x == 'D' else 4 if x == 'E' else 5 if x == 'F' else 6)                                                                               
data['default_file'] = data['cb_person_default_on_file'].apply(lambda x:0 if x == 'N' else 1)
data['home_ownership'] = data['person_home_ownership'].apply(lambda x:0 if x == 'RENT' else 1  if x == 'OWN' else 2 if x=='MORTAGE' else 3)
```

```python
data
```

```python
data = data.drop(['person_home_ownership','loan_intent', 'loan_grade' ,'cb_person_default_on_file'],axis =1)
```

```python
data
```

```python
data.columns
```

```python
X = data[['person_age', 'person_income', 'person_emp_length', 'loan_amnt',
       'loan_int_rate','loan_percent_income',
       'cb_person_cred_hist_length', 'loan_intent_duplicate',
       'loan_grade_duplicate', 'default_file', 'home_ownership']]
```

```python
Y = data[['loan_status']]
```

```python
X_test , X_train , Y_test , Y_train = train_test_split(X,Y,test_size=0.8 ,random_state=42)
```

```python
#Hpyer Parameter Tuning :-

param_grid = {
    'n_estimators' : [100,200,300],
    'max_depth' : [None , 10 ,20, 30],
    'min_samples_split' : [2,5,10],
    'min_samples_leaf' : [1,2,4]
}
```

```python
rfc = RandomForestClassifier(random_state=42)
```

```python
grid_search = GridSearchCV(estimator=rfc , param_grid=param_grid , cv=5 , n_jobs=-1 ,verbose=2)
grid_search.fit(X_train,Y_train)

best_rfc = grid_search.best_estimator_
```

```python
Y_pred = best_rfc.predict(X_test)
print(classification_report(Y_test , Y_pred))
```

```python
with open('model.pkl' , 'wb') as file:
    pickle.dump(best_rfc , file)
```
